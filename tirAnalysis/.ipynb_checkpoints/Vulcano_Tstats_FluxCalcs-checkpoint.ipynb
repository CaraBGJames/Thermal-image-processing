```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import tifffile 
# import gdal

# Configure packages and settings
# %matplotlib inline
plt.rc('text', usetex=True)
plt.rc('font', family='serif', size=12)
```

```python
# Define the location of the data
path  = '/home/david/FieldWork/Vulcano/'
fname = path + '2016-09-Vulcano/2016-09-17_nighttimeFLIR/allGeorefScenes/allSites_corrected_test.tif'

# Read the geotiff file (use tifffile library until gdal can be properly reinstalled)
Tlayer = tifffile.imread(fname)
Tlayer.shape

mask = Tlayer == 0.
Tlayer[mask] = np.nan

dempath   = path + 'DEM/06_final_20cm_model_ortho/'
demfile   = dempath + 'rgb_20cm_hillshaded.jpg'
DEMlayer  = plt.imread(demfile)
DEMextent = []
with open(dempath + 'rgb_20cm_hillshaded.jpw', 'r') as f:
    for line in f:
        DEMextent.append(float(line.strip()))
DEMextent = np.array(DEMextent)
```

```python
import rasterio as rio

TlayerDS = rio.open(fname)

meta = TlayerDS.meta
# for key in meta.keys():
#     print(key, ':', meta[key])

geotrans = meta['transform']
W, H     = meta['width'], meta['height']
nLayers  = meta['count']

def transform2extent(transform):
    return geotrans[2], geotrans[0] * W + geotrans[2], geotrans[4] * H + geotrans[5], geotrans[5]]
extent = [geotrans[2], geotrans[0] * W + geotrans[2], geotrans[4] * H + geotrans[5], geotrans[5]]

```

```python
fig, ax = plt.subplots(figsize=(12, 9))
im = ax.imshow(Tlayer, cmap=plt.cm.gray, vmin=283.15, vmax=363, extent=extent, zorder=3)

titleStr = 'Temp. stats: min = %.2f K, max = %.2f K, mean = %.2f K' % (np.nanmin(Tlayer), 
                                                                       np.nanmax(Tlayer), 
                                                                       np.nanmean(Tlayer))
h_title  = plt.title(titleStr, fontdict={'size' : 14})
h_xlabel = ax.set_xlabel('WGS84 UTM33N Eastings/[m]')
h_ylabel = ax.set_ylabel('WGS84 UTM33N Northings/[m]')

cbar = fig.colorbar(im)
h_cbarLabel = cbar.set_label('Corrected surface temperature/[K]')
cbar.minorticks_off()

fig.savefig('20160916_Vulcano_CorrectedSurfTemp_georef.pdf', dpi=300)
```
